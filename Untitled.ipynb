 4. Preprocessing

```python
URL_RE    = re.compile(r'(([\w]+:)?//)?(([\d\w]|%[A-Fa-f\d]{2,})+(:[\d\w]+)?@)?' #those are name metachracjters
                       r'([\d\w][-\d\w]{0,253}\.)+[\w]{2,4}(:\d+)?(/[\w\-\./?%&=]*)?') #and this is for removing urls 
EMAIL_RE  = re.compile(r'[a-z0-9\.\-_]+@[a-z0-9\.\-]+\.[a-z]{2,}')
HASH_RE   = re.compile(r'#(\w+)') #for hashtags
HANDLE_RE = re.compile(r'@\w+') #twitter style handles (@usename)
REPEAT_RE = re.compile(r'([A-Za-z])\1{2,}') #if a letter is repeated like (youuuuuuuuu)
NUM_RE    = re.compile(r'\d+') #any sequnce of digits
PUNC_RE   = re.compile(f"[{re.escape(string.punctuation)}]") #any punctuation character
SPACE_RE  = re.compile(r'\s+') #to remove white space

def pre_process(msg):
    msg = URL_RE.sub('', msg)                  # 1. strip URLs
    msg = EMAIL_RE.sub(' ', msg)               # 2. strip emails
    msg = HASH_RE.sub(r' \1', msg)             
    msg = HANDLE_RE.sub(' ', msg)              # 4. drop @handles
    msg = msg.encode('ascii','ignore').decode()# 5. remove non-ASCII (emojis)
    msg = msg.replace('`', "'")                # 6. normalize back-ticks
    msg = contractions.fix(msg)                # 7. expand “don’t”→“do not”
    msg = REPEAT_RE.sub(r'\1', msg)            # 8. “soooo”→“so”
    msg = NUM_RE.sub(' ', msg)                 # 9. remove digits
    msg = PUNC_RE.sub(' ', msg)                # 10. remove punctuation
    msg = msg.lower().strip()                  # 11. lowercase + trim edges
    return SPACE_RE.sub(' ', msg)              # 12. collapse extra spaces
    
df['clean'] = df['text'].map(pre_process)
print("\n3) Sample cleaned text:")
display(df[['text','clean']].head())
```

 5. Bow Tokenize → Stop-words → Lemmatize

```python
tok        = TreebankWordTokenizer()
stop_words = set(stopwords.words('english'))
lemmatizer = WordNetLemmatizer()

df['tokens']        = df['clean'].map(tok.tokenize)
df = df[df['tokens'].map(len)>0]
df['tokens_nostop'] = df['tokens'].map(lambda toks: [t for t in toks if t not in stop_words]) #remove stop words 
df['tokens_lem']    = df['tokens_nostop'].map(lambda toks: [lemmatizer.lemmatize(t) for t in toks]) #

print("\n4) Sample tokens:")
display(df[['clean','tokens_lem']].head())
```

 6. EDA: Most Common Words

```python
ct_all = Counter(w for toks in df['tokens_lem'] for w in toks)
top20  = dict(ct_all.most_common(20))

plt.figure(figsize=(6,3))
plt.bar(top20.keys(), top20.values())
plt.xticks(rotation=45)
plt.title("Top-20 Tokens (All)")
plt.tight_layout()
plt.show()
```

```python
wc = WordCloud(width=400, height=200).generate_from_frequencies(ct_all)
plt.figure(figsize=(6,3))
plt.imshow(wc); plt.axis('off')
plt.title("Word Cloud (All)")
plt.tight_layout()
plt.show()
```

 7. Prepare X, y & Encode Labels

```python
df['sentence'] = df['tokens_lem'].map(' '.join)
X = df['sentence']
y = df['label']
le = LabelEncoder() 
y_enc = le.fit_transform(y)
print("\n5) Label encoding:", dict(zip(le.classes_, le.transform(le.classes_))), "\n")
```

8. TF-IDF & Train/Test Split

```python
tfidf    = TfidfVectorizer(min_df=5, ngram_range=(1,2)) #min df=5 ignore any word that appear only 5 times, single words and two words
X_vec    = tfidf.fit_transform(X)
X_tr,X_te,y_tr,y_te = train_test_split(X_vec, y_enc, stratify=y_enc, test_size=0.2, random_state=42)
print("6) Shapes — X_tr:", X_tr.shape, "X_te:", X_te.shape, "y_tr:", np.bincount(y_tr), "y_te:", np.bincount(y_te), "\n")
```

9. SMOTE & Show Resampling

```python
print("7) Before SMOTE:", np.bincount(y_tr))
sm    = SMOTE(random_state=42)
X_res, y_res = sm.fit_resample(X_tr, y_tr)
print("   After SMOTE: ", np.bincount(y_res))

plt.figure(figsize=(5,3))
sns.countplot(x=y_res)
plt.title("Training Labels After SMOTE")
plt.xticks([0,1], le.inverse_transform([0,1]))
plt.tight_layout()
plt.show()
```

 10. Train & Evaluate & Compare Models

```python
models = {
    'MultinomialNB'      : MultinomialNB(),
    'LogisticRegression' : LogisticRegression(max_iter=1000),
    'DecisionTree'       : DecisionTreeClassifier(max_depth=10, random_state=42),
    'RandomForest'       : RandomForestClassifier(n_estimators=100, max_depth=10, random_state=42),
    'GradientBoosting'   : GradientBoostingClassifier(learning_rate=0.1, n_estimators=100, random_state=42),
    'SVM'                : SVC(kernel='rbf', probability=True),
    'KNN'                : KNeighborsClassifier(n_neighbors=5)
}

results = []
for name, mdl in models.items():
    mdl.fit(X_res, y_res)
    pred = mdl.predict(X_te)
    results.append({
        'Model'     : name,
        'Accuracy'  : accuracy_score(y_te, pred),
        'Precision' : precision_score(y_te, pred, pos_label=1),
        'Recall'    : recall_score(y_te, pred, pos_label=1),
        'F1 Score'  : f1_score(y_te, pred, pos_label=1)
    })
```

```python
results_df = pd.DataFrame(results).sort_values('Accuracy', ascending=False).reset_index(drop=True)
print("8) Model comparison:\n", results_df, "\n")
```

```python
res_m = results_df.melt(id_vars='Model', var_name='Metric', value_name='Score')
plt.figure(figsize=(8,4))
sns.barplot(data=res_m, x='Model', y='Score', hue='Metric')
plt.xticks(rotation=45)
plt.title("Model Performance Comparison")
plt.tight_layout()
plt.show()
```

11. Confusion Matrix for Best Model

```python
best = results_df.loc[results_df['Accuracy'].idxmax(), 'Model']
mdl  = models[best]
y_pred = mdl.predict(X_te)

cm = confusion_matrix(y_te, y_pred)
plt.figure(figsize=(4,3))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues',
            xticklabels=le.classes_, yticklabels=le.classes_)
plt.xlabel("Predicted"); plt.ylabel("Actual")
plt.title(f"Confusion Matrix ({best})")
plt.tight_layout()
plt.show()
```
